```python
# Sample text
text = "Apple is looking at buying U.K. startup for $1 billion"

# Process the text
doc = nlp(text)
```

```python
# Process the words
doc = nlp(" ".join(list_of_words))
doc
```

```python
# tokenization & POS tagging
lexical_info = []

for token in doc:
    print(f"Token: {token.text}, POS: {token.pos_}, Tag: {token.tag_}, Lemma: {token.lemma_}, Entity: {token.ent_type_}, Stop Word: {token.is_stop}, Length: {token.__len__()}")
    tmp_dict = {}
    tmp_dict['token'] = token.text
    tmp_dict['pos_'] = token.pos_
    tmp_dict['tag_'] = token.tag_
    tmp_dict['lemma_'] = token.lemma_
    tmp_dict['ent_type_'] = token.ent_type_
    tmp_dict['is_stop'] = token.is_stop
    tmp_dict['length'] = token.__len__()
    lexical_info.append(tmp_dict)
```

```python
lexical_df = pd.DataFrame(lexical_info)
lexical_df.to_csv('lexical_df.csv')

lexical_df_wo_stop = lexical_df[lexical_df['is_stop'] == False]
lexical_df_wo_stop.head()
token_list = lexical_df_wo_stop['token'].tolist()
token_list = [x for x in token_list if len(x) > 2]
```

```python
word_freq = Counter(token_list)
common_words = word_freq.most_common(25)
first=[]
second=[]
for i in common_words:
    first.append(i[0])
    second.append(i[1])
plt.figure(figsize=(8, 8))

plt.bar(first, second,  color='orange')
plt.xticks(range(len(first)), first, rotation='vertical')
plt.title('Frequency of toxic words')
plt.savefig('word_freq')
plt.show()
```
